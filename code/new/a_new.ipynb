```python
import os

# Function to remove double quotes from lines in a file
def remove_double_quotes_from_file(file_path):
    # Read lines from the input file
    with open(file_path, 'r') as f:
        lines = f.readlines()

    # Remove double quotes from each line
    cleaned_lines = [line.replace('"', "") for line in lines]

    # Write cleaned lines back to the same file
    with open(file_path, 'w') as f:
        f.writelines(cleaned_lines)

    print(f"Processed {file_path}. Double quotes removed.")

# Root directory containing the subdirectories with files to process
root_directory = 'input_files'
subdirectories = ['left', 'right', 'center']

# Iterate over each subdirectory in the root directory
for subdir in subdirectories:
    folder = os.path.join(root_directory, subdir)
    # Iterate over each file in the current subdirectory
    for filename in os.listdir(folder):
        file_path = os.path.join(folder, filename)
        
        # Check if the file is a regular file (not a directory)
        if os.path.isfile(file_path):
            # Remove double quotes from the current file
            remove_double_quotes_from_file(file_path)
```
